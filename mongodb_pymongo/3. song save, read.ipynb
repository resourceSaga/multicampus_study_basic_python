# Song info save and read
- data/songs.json read
- MongoDB connect
- insert_many() 등록, find() 조회

```python
import json
import pymongo

# json file read
with open('data/songs.json', 'r', encoding='utf8') as file:
    contents = file.read()
    json_data = json.loads(contents)
```

```python
print(type(json_data))
```

```python
import pymongo

conn = pymongo.MongoClient()
print(conn)
song_db = conn.song_db
print(song_db)
song_col = song_db.song_col
print(song_col)
```

```python
song_col.insert_many(json_data)
song_col.count_documents({})
```

```python
for song in song_col.find({}, {"_id":0, "가사":0}):
    print(song)
```

```python
for genre in song_col.aggregate([{"$group":{"_id":"$장르"}}]):
    print(genre)
```

```python
# genre 댄스인 곡명, 가수, 앨범, 발매일
for idx, info in enumerate (song_col.find({"장르":"댄스"},{"_id":0,"곡명":1,"가수":1,"앨범":1,"발매일":1})):
    print(idx, info)
```

```python
# 가수가 방탄소년단, 폴킴의 곡명, 가수, 앨범, 발매일
for info in song_col.find({"$or":[{"가수":"방탄소년단"},{"가수":"폴킴"}]},{"_id":0,"곡명":1,"가수":1,"앨범":1,"발매일":1}):    
    print(info)
```

```python
# 앨범 이름 중에서 OST 가 포함된 노래의 곡명, 가수, 앨범, 발매일
for info in song_col.find({"앨범":{"$regex":"OST"}},{"_id":0,"곡명":1,"가수":1,"앨범":1,"발매일":1}):
    print(info)
```

```python
# 가수별 group by
for artist in song_col.aggregate([
    {"$group":
     {"_id":"$가수",
     "count":{"$sum":1}
     }     
    },
    {
        "$sort":{"count":-1}
    }
]):
    print(artist)
```
