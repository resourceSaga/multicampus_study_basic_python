## Read_csv() 함수
- csv file 읽기
- head(), tail() 함수로 일부 데이터 가져오기
- columns, index, shape, info() 로 meta 정보 확인
- describe(), max(), min(), mean() 집계함수
- 컬럼명을 변경하고, 인덱스를 바꾸기
- 컬럼 한개는 pandas 고유 Series 객체, 2개 이상은 DataFrame

```python
import pandas as pd

data = pd.read_csv('data/data_draw_korea.csv')
print(type(data))
data.head() # head default argument is 5
data.tail() # tail default argurment is 5

print(data.columns) # property 
print(data.index) # index
print(data.shape) # return number of (row, column)
print(data.info()) # dataframe info
```

```python
# describe() 집계함수 
data.describe()
```

```python
# 인구수 컬럼값의 최대, 최소, 평균
type(data['인구수'])

print('min', data['인구수'].min())
print('max', data['인구수'].max())
print('avg', data['인구수'].mean())
print('편차', data['인구수'].std())
```

```python
# 컬럼명 변경
data = data.rename(columns={'Unnamed: 0':'seq'})
data.head(2)
```

```python
# seq 컬럼을 index로 변경하기
data = data.set_index('seq')
data.head(2)
```

## 특정 행과 열을 선택하기

- loc[] 사용, iloc[] #location의 약자   #loc[row,column]
- 특정 열을 선택, slicing으로 열을 선택
- 특정 행을 선택, slicing으로 행을 선택
- boolean indexing - 특정 조건식을 만족하는 행을 선택

```python
# 인구수, 행정구역 2개의 컬럼 선택
data.loc[:, ['인구수','행정구역']].head()

print(type(data.loc[:,'인구수']))

# 인구수, 행정구역, 광역시도 3개의 컬럼 선택
data.loc[:, ['인구수','행정구역','광역시도']].head()
```

```python
# 인구수 컬럼부터 면적 컬럼까지
data.loc[:,'인구수':'면적']
```

```python
col_list = ['인구수', '행정구역', '광역시도']
data.loc[:, col_list].head()
```

```python
# 1개 행만 선택
data.loc[0,:]
```

```python
# 여러개의 특정행
data.loc[[9,0,4],:]
```

```python
# 여러개의 행 (구간-slicing)
data.loc[0:3,:]
# data.loc[0:10:2,:]
```

```python
# 1. 20개의 행, '행정구역', '광역시도' 가져오기
data.loc[0:19,['행정구역', '광역시도']]
# 2. 10, 20, 30행, '인구수' 부터 '면적' 까지
data.loc[[10, 20, 30],'인구수':'면적']
```

```python
# iloc[] 사용
data.iloc[0:3,0:3]
```

```python
# 광역시도 이름 확인 (중복된 이름 빼고)
data['광역시도'].unique()
```

```python
# group by count('광역시도')
data['광역시도'].value_counts()
```

```python
data['광역시도'] == '서울특별시'
```

```python
# loc 에 조건을 줘서 sorting ( boolean indexing)
data.loc[data['광역시도'] == '서울특별시',:]
```

```python
# 부산광역시의 행정구역, 인구수 컬럼을 가져오기
data.loc[data['광역시도'] == '부산광역시',['행정구역','인구수']]
```

```python
# 인구수가 평균보다 낮은 지역
pop_mean_value = data['인구수'].mean()
data.loc[data['인구수'] < pop_mean_value , '행정구역']

# 면적이 평균보다 작은 곳
area_mean_value = data['면적'].mean()
data.loc[data['면적'] < area_mean_value , :]
```

```python
# 경기도에서 인구수가 평균보다 높은 행정구역
data.loc[(data['광역시도'] == '경기도') & (data['인구수']> pop_mean_value), '행정구역']
```

```python
# 전체에서 표준 편차 인구수 보다 낮거나, 평균 면적 보다 높은 지역은?
data.loc[(data['인구수'] < data['인구수'].std()) | (data['면적'] > data['면적'].mean()) ,['광역시도','행정구역','인구수','면적']]
```

```python
def get_pop_max(sido):
    return data.loc[data['광역시도'] == sido, '인구수'].max()
```

```python
get_pop_max('서울특별시')
```

```python
data.loc[data['인구수'] == get_pop_max('서울특별시')]
```

```python
def get_pop_max_df(sido):
    return data.loc[data['인구수'] == get_pop_max(sido) , ['광역시도', '행정구역', '인구수']]
```

```python
type(data['광역시도'].unique())
```

```python
for sido in data['광역시도'].unique():
    get_pop_max_df(sido)
```

```python
# 새로운 DataFrame 생성하기
max_pop_df = pd.DataFrame(columns=['광역시도', '행정구역', '인구수'])
for sido in data['광역시도'].unique():
    max_pop_df = max_pop_df.append(get_pop_max_df(sido))
```

```python
# reset_index(drop=false) - 인덱스 조정 
# 인덱스를 변경하면서 기존 인덱스 값을 컬럼으로변경
# drop default false. True 시 드랍
max_pop_df = max_pop_df.reset_index(drop=True)
```

```python
# 인덱스를 1 부터 시작하는 방법
import numpy as np
max_pop_df.index = np.arange(1, len(max_pop_df)+1)
print(max_pop_df.index)
max_pop_df
```

```python
def get_area_max(sido):
    return data.loc[data['광역시도'] == sido, '면적'].max()

def get_area_max_df(sido):
    return data.loc[data['면적'] == get_area_max(sido) , ['광역시도', '행정구역', '면적']]

# 새로운 DeataFrame 생성하기2
max_area_df = pd.DataFrame(columns=['광역시도', '행정구역', '면적'])

for sido in data['광역시도'].unique():
    max_area_df = max_area_df.append(get_area_max_df(sido))
```

```python
max_area_df = max_area_df.reset_index(drop=True)
```

```python
max_area_df
```

```python
# 값을 원하는 조건으로 정렬
# shift + tab : 함수 help
max_pop_df.sort_values(by='인구수',ascending=False)
```

- 1에 가까울 수록 서로 관련성이 있음
- corr() 함수를 사용

```python
# 전체 면적고 인구수의 상관관계 1
data['면적'].corr(data['인구수'])
```

```python
# 서울특별시의 면적과 인구수의 상관관계
seoul_df = data.loc[data['광역시도'] == '서울특별시']
```

```python
seoul_df['면적'].corr(seoul_df['인구수'])
```

## Group by
- Series 객체의 groupby()
- DataFrame 객체의 groupby()

```python
data['광역시도'].value_counts()
```

```python
# 광역시도별 인구수의 합계
# ~ 별에 해당하는 컬럼은 groupby() 함수의 파라미터로 전달한다.
pop_data = data['인구수'].groupby(data['광역시도']).sum().sort_values()
```

```python
# 광역시도별 면적의 합계 - descending
data['면적'].groupby(data['광역시도']).sum().sort_values()
```

```python
# DataFrame 객체의 groupby
# 광역시도별 인구의 합계
data.groupby('광역시도').sum()
```

```python
data.groupby('광역시도')['인구수'].sum()
```

```python
data.groupby('광역시도')['면적'].sum()
```

```python
# 광역시도별, 행정구역별 인구수
grouping_data = data.groupby(['광역시도', '행정구역'])['인구수'].sum()
type(grouping_data)
```

```python
# excel 파일 저장
grouping_data.to_excel('data/광역시도행정구역별인구수.xlsx', sheet_name='인구수')
```
