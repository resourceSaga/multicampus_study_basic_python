## Read_csv() 함수
- csv file 읽기
- head(), tail() 함수로 일부 데이터 가져오기
- columns, index, shape, info() 로 meta 정보 확인
- describe(), max(), min(), mean() 집계함수
- 컬럼명을 변경하고, 인덱스를 바꾸기
- 컬럼 한개는 pandas 고유 Series 객체, 2개 이상은 DataFrame

```python
import pandas as pd

data = pd.read_csv('data/data_draw_korea.csv')
print(type(data))
data.head() # head default argument is 5
data.tail() # tail default argurment is 5

print(data.columns) # property 
print(data.index) # index
print(data.shape) # return number of (row, column)
print(data.info()) # dataframe info
```

```python
# describe() 집계함수 
data.describe()
```

```python
# 인구수 컬럼값의 최대, 최소, 평균
type(data['인구수'])

print('min', data['인구수'].min())
print('max', data['인구수'].max())
print('avg', data['인구수'].mean())
```

```python
# 컬럼명 변경
data = data.rename(columns={'Unnamed: 0':'seq'})
data.head(2)
```

```python
# seq 컬럼을 index로 변경하기
data = data.set_index('seq')
data.head(2)
```

## 특정 행과 열을 선택하기

- loc[] 사용, iloc[] #location의 약자   #loc[row,column]
- 특정 열을 선택, slicing으로 열을 선택
- 특정 행을 선택, slicing으로 행을 선택
- boolean indexing - 특정 조건식을 만족하는 행을 선택

```python
# 인구수, 행정구역 2개의 컬럼 선택
data.loc[:, ['인구수','행정구역']].head()

print(type(data.loc[:,'인구수']))

# 인구수, 행정구역, 광역시도 3개의 컬럼 선택
data.loc[:, ['인구수','행정구역','광역시도']].head()
```

```python
# 인구수 컬럼부터 면적 컬럼까지
data.loc[:,'인구수':'면적']
```

```python
col_list = ['인구수', '행정구역', '광역시도']
data.loc[:, col_list].head()
```

```python
# 1개 행만 선택
data.loc[0,:]
```

```python
# 여러개의 특정행
data.loc[[9,0,4],:]
```

```python
# 여러개의 행 (구간-slicing)
data.loc[0:3,:]
# data.loc[0:10:2,:]
```

```python
# 1. 20개의 행, '행정구역', '광역시도' 가져오기
data.loc[0:19,['행정구역', '광역시도']]
# 2. 10, 20, 30행, '인구수' 부터 '면적' 까지
data.loc[[10, 20, 30],'인구수':'면적']
```
